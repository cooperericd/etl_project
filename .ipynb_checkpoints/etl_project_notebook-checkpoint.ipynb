```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
airbnb_file = "AB_NYC_2019.csv"
```

```python
airbnb_df = pd.read_csv(airbnb_file)
```

```python
airbnb_df.head()
```

```python
airbnb_df = airbnb_df.rename(columns={"host_id": "host_number", "host_name": "host_title"})
```

```python
# Create a filtered dataframe from specific columns
host_cols = ["host_number", "host_title"]
host_df = airbnb_df[host_cols].copy()

# Clean the data by dropping duplicates and setting the index
host_df.drop_duplicates("host_number", inplace=True)
host_df.set_index("host_number", inplace=True)

host_df.head()
```

```python
# Create a filtered dataframe from specific columns
neighbourhood_group_cols = ["neighbourhood_group"]
neighbourhood_group_transformed = airbnb_df[neighbourhood_group_cols].copy()

# Rename the column headers
neighbourhood_group_transformed = neighbourhood_group_transformed.rename(columns={"neighbourhood_group":"borough"})
# Clean the data by dropping duplicates and setting the index
neighbourhood_group_transformed.drop_duplicates("borough", inplace=True)
neighbourhood_group_transformed.set_index("borough", inplace=True)
neighbourhood_group_transformed = neighbourhood_group_transformed.reset_index()
neighbourhood_group_transformed.index= neighbourhood_group_transformed.index + 1
neighbourhood_group_transformed = neighbourhood_group_transformed.reset_index()
neighbourhood_group_transformed = neighbourhood_group_transformed.rename(columns={"index": "borough_id"})

neighbourhood_group_transformed.head()
```

```python
neighbourhood_df =  airbnb_df[["neighbourhood"]]
neighbourhood_df = neighbourhood_df.dropna(how='any')
neighbourhood_df_unique = neighbourhood_df.drop_duplicates()
neighbourhood_df_unique.set_index("neighbourhood", inplace=True)
neighbourhood_df_unique = neighbourhood_df_unique.reset_index()
neighbourhood_df_unique.index = neighbourhood_df_unique.index+1
neighbourhood_df_unique = neighbourhood_df_unique.reset_index()
neighbourhood_df_unique = neighbourhood_df_unique.rename(columns={"index":"neighbourhood_id"})
neighbourhood_df_unique.head()
```

```python
room_cols=["host_number", "room_type"]
room_type=airbnb_df[room_cols].copy()
room_type.drop_duplicates("room_type", inplace=True)
room_type=room_type.drop(columns=["host_number"])
room_type.index=range(1,4)
room_type=room_type.reset_index()
room_type=room_type.rename(columns={"index":"room_type_id"})
room_type
```

```python
listing_df = pd.DataFrame.merge(airbnb_df, neighbourhood_df_unique, how='inner', left_on='neighbourhood', right_on='neighbourhood')
```

```python
listing_df_rev = pd.DataFrame.merge(listing_df, room_type, how='inner', left_on='room_type', right_on='room_type')
```

```python
listing_df_rev.head()
```

```python
listing_df_copy = listing_df_rev.copy()
listing_df_reduced = listing_df_copy[['id', 'name', 'host_number', 'latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'last_review', 'reviews_per_month', 'availability_365', 'neighbourhood_id', 'room_type_id']]
```

```python
listing_df_reduced.head()
```

```python
pw = input("Password: ")
connection_string = "postgres:"+str(pw)+"@localhost:5432/airbnb_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
host_df.to_sql(name='host', con=engine, if_exists='append', index=True)
```

```python
neighbourhood_group_transformed.to_sql(name='borough', con=engine, if_exists='append', index=False)
```

```python
neighbourhood_df_unique.to_sql(name='neighbourhood', con=engine, if_exists='append', index=False)
```

```python
room_type.to_sql(name='room_type', con=engine, if_exists='append', index=False)
```

```python
listing_df_reduced.to_sql(name='listing', con=engine, if_exists='append', index=False)
```
