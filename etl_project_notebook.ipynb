```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
airbnb_file = "AB_NYC_2019.csv"
```

```python
airbnb_df = pd.read_csv(airbnb_file)
```

```python
airbnb_df.head()
```

```python
# Create a filtered dataframe from specific columns
host_cols = ["host_id", "host_name"]
host_df = airbnb_df[host_cols].copy()

# Rename the column headers
host_df = host_df.rename(columns={"host_id": "id"})

# Clean the data by dropping duplicates and setting the index
host_df.drop_duplicates("id", inplace=True)
host_df.set_index("id", inplace=True)

host_df.head()
```

```python
(airbnb_df['last_review'][2])
```

```python
# Create a filtered dataframe from specific columns
neighbourhood_group_cols = ["neighbourhood_group"]
neighbourhood_group_transformed = airbnb_df[neighbourhood_group_cols].copy()

# Rename the column headers
neighbourhood_group_transformed = neighbourhood_group_transformed.rename(columns={"neighbourhood_group":"borough"})
# Clean the data by dropping duplicates and setting the index
neighbourhood_group_transformed.drop_duplicates("borough", inplace=True)
neighbourhood_group_transformed.set_index("borough", inplace=True)
neighbourhood_group_transformed = neighbourhood_group_transformed.reset_index()
neighbourhood_group_transformed.index= neighbourhood_group_transformed.index + 1
neighbourhood_group_transformed = neighbourhood_group_transformed.reset_index()
neighbourhood_group_transformed = neighbourhood_group_transformed.rename(columns={"index": "borough_id"})

neighbourhood_group_transformed.head()
```

```python
neighbourhood_df =  airbnb_df[["neighbourhood"]]
neighbourhood_df = neighbourhood_df.dropna(how='any')
neighbourhood_df_unique = neighbourhood_df.drop_duplicates()
neighbourhood_df_unique.set_index("neighbourhood", inplace=True)
neighbourhood_df_unique = neighbourhood_df_unique.reset_index()
neighbourhood_df_unique.index = neighbourhood_df_unique.index+1
neighbourhood_df_unique = neighbourhood_df_unique.reset_index()
neighbourhood_df_unique = neighbourhood_df_unique.rename(columns={"index":"neighbourhood_id"})
neighbourhood_df_unique.head()
```

```python
room_cols=["host_id", "room_type"]
room_type=airbnb_df[room_cols].copy()
room_type.drop_duplicates("room_type", inplace=True)
room_type=room_type.drop(columns=["host_id"])
room_type.index=range(1,4)
room_type
```

```python
listing_df = pd.DataFrame.merge(airbnb_df, neighbourhood_group_transformed, how='inner', left_on="neighbourhood_group", right_on="borough")
```

```python
listing_df_rev = pd.DataFrame.merge(listing_df, neighbourhood_df_unique, how='inner', left_on='neighbourhood', right_on='neighbourhood')
```
